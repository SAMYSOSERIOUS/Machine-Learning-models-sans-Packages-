Importing libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

loading data with keras

```python
(x_train, y_train),(x_test, y_test) = keras.datasets.mnist.load_data()
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

one image's shape

```python
len(x_train)
len(x_test)
x_train[0].shape
plt.matshow(x_train[0])
```

dataset normalization

```python
# Normalizing the dataset
x_train = x_train/255
x_test = x_test/255

# Flatting the dataset in order
# to compute for model building
x_train_flatten = x_train.reshape(len(x_train), 28*28)
x_test_flatten = x_test.reshape(len(x_test), 28*28)
```

Building a Neural Network

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import models as keras_models
from tensorflow.keras import layers as keras_layers
model = keras_models.Sequential([
    keras_layers.Dense(100, input_shape=(784,), activation='sigmoid'),])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])

model.fit(x_train_flatten, y_train, epochs=5)
```

model evaluation

```python
model.evaluate(x_test_flatten, y_test)
```

```python

import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Training the model and storing the history
history = model.fit(x_train_flatten, y_train, epochs=5)

# Plot the loss graph
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()

# model evaluation
model.evaluate(x_test_flatten, y_test)
```
